# Laden des Datensatzes

```python
import pandas as pd
from sklearn import datasets
import matplotlib.pyplot as plt

iris = datasets.load_iris()
iris_df = pd.DataFrame(iris.data)
iris_df['class']=iris.target_names[iris.target ]
iris_df.columns=['sepal_len', 'sepal_wid', 'petal_len', 'petal_wid', 'class']
```

# Single Numerical Feature

## Swarmplot

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set_theme()

fig, ax = plt.subplots( figsize=(9,6))
ax.set_title("Iris Swarmplot - petal-width")
ax = sns.swarmplot(data = iris_df,  x = "petal_wid")
```

## Histogramm

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set()
sns.histplot(data = iris_df, 
             x ="petal_wid", 
             bins=10,  # try bins="auto" und bins=10
             kde = False)
```

## Violinplot

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set_theme()

fig, ax = plt.subplots( figsize=(9,6))
ax.set_title("Iris Swarmplot - petal-width")
ax = sns.violinplot(data = iris_df,  x = "petal_wid", inner="quartiles")
```

## Boxplot

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set_theme()

fig, ax = plt.subplots( figsize=(9,6))
ax.set_title("Iris Swarmplot - petal-width")
ax = sns.boxplot(data = iris_df,  x = "petal_wid")
```

# Relation-Plots

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set()
sns.set_context("notebook", font_scale=1.2)
sns.relplot(data=iris_df, x="petal_len", y="sepal_wid", hue = "class", height=7, s = 50) # try style="class"
```

# Pairplots

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set()
sns.set_context("notebook", font_scale=1.3)

sns.pairplot(iris_df[['sepal_len', 'sepal_wid', 'petal_len', 'petal_wid', 'class']],
             hue="class", diag_kind="kde")
```
